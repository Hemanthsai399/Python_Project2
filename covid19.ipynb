```python
top_states_confirmed=covid.groupby(['State/UnionTerritory'])['Confirmed'].sum().sort_values(ascending=False).head(10)
top_states_confirmed
```

```python
plt.plot(top_states_confirmed);
```

```python
top_states_deaths=covid.groupby(['State/UnionTerritory'])['Deaths'].sum().sort_values(ascending=False).head(10)
top_states_deaths
```

```python
plt.plot(top_states_deaths);
```

```python
top_states_cured=covid.groupby(['State/UnionTerritory'])['Cured'].sum().sort_values(ascending=False).head(10)
top_states_cured
```

```python
plt.plot(top_states_cured);
```

```python
sns.set(rc={'figure.figsize':(25,8)})
```

```python
#now line graph to show top deaths and cured
plt.plot(top_states_deaths,top_states_cured)
plt.xlabel('top_states_deaths')
plt.ylabel('top_states_cured')
plt.title('Top ststes deaths and cured');
```
